```python
import pandas as pd
import ast
```

```python
#select dataset train vs test

data_split= 'test'
```

```python
df_metadata = pd.read_excel("./annotation/AR_metadata.xlsx", sheet_name='AR')


```

```python
df_metadata["list_animal"]= df_metadata["list_animal"].apply(ast.literal_eval)
df_metadata["list_animal_parent_class"]= df_metadata["list_animal_parent_class"].apply(ast.literal_eval)
df_metadata["list_animal_action"]= df_metadata["list_animal_action"].apply(ast.literal_eval)

df_metadata
```

```python
df_metadata= df_metadata[df_metadata["type"]==data_split]
len(df_metadata)
```

```python
def format_action(animal_action):
    
    return f"{animal_action[0]} is {animal_action[1]}"

df_metadata['list_animal_action'] = df_metadata['list_animal_action'].apply(lambda x: [format_action(t) for t in x])
```

```python
df_metadata
```

```python
dist_animals_inner_dict={}
dist_animals_inner_dict["from"]="human"
dist_animals_inner_dict["value"]="Below is a video of animals performing some actions. Memorize the content and answer my question after this video. The video starts.\n<image>\nThe video ends.\n Tell me what are the distinct animal species in this video. If you notice a species twice only mention it once"


act_animals_inner_dict={}
act_animals_inner_dict["from"]="human"
act_animals_inner_dict["value"]="Tell me for each animal what action it performs. Give independent unique sentences for each animal for each action."

gen_animals_inner_dict={}
gen_animals_inner_dict["from"]="human"
gen_animals_inner_dict["value"]="Tell me what are the distinct animal classes in this video. If you notice a the same class twice only mention it once"

output_list= []
for index, row in df_metadata.iterrows():    
    insert={}
    insert["id"]=row["video_id"] 
    insert["video"]=f"action_recognition/video/{row["video_id"]}.mp4"

    insert["conversations"]=[]

    second_inner_dict={}
    second_inner_dict["from"]="gpt"
    second_inner_dict["value"]='The distinct species I see are '+', '.join(list(set(row["list_animal"])))
    insert["conversations"].append(dist_animals_inner_dict)
    insert["conversations"].append(second_inner_dict)


    second_inner_dict={}
    second_inner_dict["from"]="gpt"
    second_inner_dict["value"]='The distinct categories I see are '+ ', '.join(list(set(row["list_animal_parent_class"])))
    insert["conversations"].append(gen_animals_inner_dict)
    insert["conversations"].append(second_inner_dict)


    second_inner_dict={}
    second_inner_dict["from"]="gpt"
    second_inner_dict["value"]='. '.join(list(set(row["list_animal_action"])))
    insert["conversations"].append(act_animals_inner_dict)
    insert["conversations"].append(second_inner_dict)


    output_list.append(insert)


```

```python
import json
with open(f"{data_split}_annotation_animals_action_recognition.json", "w") as final:
    json.dump(output_list, final)
```
