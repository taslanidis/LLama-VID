```python
import pandas as pd
```

```python
from collections import defaultdict

data_dict= defaultdict(str)
```

```python
#select dataset
dataset='train'
```

```python
file_path = f"./grounding/{dataset}.txt"

with open(file_path, 'r') as file:
        lines = file.readlines()

data = []
for l in lines:
    code_text = l.strip().split('##')
    if len(code_text) == 2:
            code_and_times, description = code_text
            
            first_space_index = code_and_times.find(' ')
            code = code_and_times[:first_space_index]
            
            if data_dict[code]:
                data_dict[code] += ' '
            data_dict[code] += description.strip()
            
            #data.append({
            ##    'Code': code.strip(),
             #   'Description': description.strip()
            #})
    
    # Create a DataFrame from the parsed data
    df = pd.DataFrame(data)
    
        
```

```python
data_dict
```

```python
first_inner_dict={}
first_inner_dict["from"]="human"
first_inner_dict["value"]="For each animal find the action it performs in the video <image>.\n Provide independent sentences for each animal and their corresponding action in the video."

output_list=[]
for key in data_dict:
    insert={}
    insert["id"]=key
    insert["video"]=f"animal-grounding/dataset/{key}.mp4"
    insert["conversations"]=[]
    second_inner_dict={}
    second_inner_dict["from"]="gpt"
    second_inner_dict["value"]=data_dict[key]
    insert["conversations"].append(first_inner_dict)
    insert["conversations"].append(second_inner_dict)
    output_list.append(insert)
```

```python
import json
with open(f"{dataset}_tony_grounding_animals.json", "w") as final:
    json.dump(output_list, final)
```
