```python
import pandas as pd
```

```python
from collections import defaultdict

data_dict= defaultdict(str)
```

```python
#select dataset
dataset='test'
```

```python
file_path = f"/home/scur0405/LLaMA-VID/data/LLaMA-VID-Eval/animal-kingdom/grounding/{dataset}.txt"

with open(file_path, 'r') as file:
        lines = file.readlines()

data = []
for l in lines:
    code_text = l.strip().split('##')
    if len(code_text) == 2:
            code_and_times, description = code_text
            
            first_space_index = code_and_times.find(' ')
            code = code_and_times[:first_space_index]
            
            if data_dict[code]:
                data_dict[code] += ' '
            data_dict[code] += description.strip()
            
            #data.append({
            ##    'Code': code.strip(),
             #   'Description': description.strip()
            #})
    
    # Create a DataFrame from the parsed data
    df = pd.DataFrame(data)
    
        
```

```python
output_list=[]
for idx, key in enumerate(data_dict):
    if idx < (len(data_dict) // 2):
        insert={}
        insert["question_id"]= idx + 1
        insert["video_name"]=f"{key}.mp4"
        insert["question"]="Find the action each animal performs in the video. Provide independent sentences for each animal and their corresponding action in the video. Do not repeat sentences. Pay careful attention, as some videos only have a single animal through the video."
        insert["answer"]= data_dict[key]
        output_list.append(insert)
```
